```python
X_train = pd.read_csv('/content/drive/MyDrive/autoML/train.csv')
X_test = pd.read_csv('/content/drive/MyDrive/autoML/test.csv', index_col=0)
```

```python
from pycaret.regression import *
```

```python
regressor = setup(X_train, 
                  remove_outliers=True, 
                  normalize=True,fold=4, 
                  target='SalePrice', 
                  feature_selection=True,
                  feature_selection_threshold= 0.7,
                  ignore_low_variance=True,
                  silent=True,
                  outliers_threshold = 0.05,
                   multicollinearity_threshold = 0.9
                    )
```

```python
Melhor_modelo = compare_models(sort= 'RMSE')
```

```python
br_regressor = create_model('br', fold=4)
```

```python
#tuned_regressor = tune_model(br_regressor,fold=4)
```

```python
predictions = predict_model(br_regressor, data=X_test)
```

```python
predictions.head()
```

```python
submit = pd.DataFrame({'Id': X_test.index,
                      'SalePrice': predictions['Label']})
submit.set_index('Id', inplace=True)
submit.to_csv('auto_ml_sub.csv')
```
