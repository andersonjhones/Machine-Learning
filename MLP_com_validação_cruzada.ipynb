#MLP (Multilayer perceptron) com validação cruzada


No presente trabalho foi utilizada a base de dados Diabetic Retinopathy Debrecen Data Set do Center for Machine Learning and Intelligent Systems. Essa base contém recursos extraídos do conjunto de imagens de Messidor para prever se uma imagem contém sinais de retinopatia diabética ou não. 
A extração de características das imagens são dadas da seguinte maneira: 

**Atributos do banco de dados**

0) O resultado binário da avaliação da qualidade. 0 = má qualidade 1 = qualidade suficiente.
1) O resultado binário da pré-triagem, onde 1 indica anormalidade retiniana severa e 0 sua falta.
2-7) Os resultados da detecção de MA. Cada valor de característica representa o
número de MAs encontrados nos níveis de confiança alfa = 0,5,. . . , 1, respectivamente.
8-15) contêm as mesmas informações que 2-7) para exsudatos. No entanto,
como os exsudatos são representados por um conjunto de pontos em vez do número de
pixels que constroem as lesões, esses recursos são normalizados dividindo o
número de lesões com o diâmetro da ROI para compensar os diferentes
tamanhos de imagem .
16) A distância euclidiana do centro de
a mácula e o centro do disco óptico para fornecer informações importantes
sobre a condição do paciente. Esse recurso
também é normalizado com o diâmetro da ROI.
17) O diâmetro do disco óptico.
18) O resultado binário da classificação baseada em AM / FM.
19) Etiqueta da classe. 1 = contém sinais de DR (etiqueta acumulativa para as classes 1, 2, 3 de Messidor), 0 = nenhum sinal de DR.





Fonte: Retinopathy Debrecen Data Set [link text](https://archive.ics.uci.edu/ml/datasets/Diabetic+Retinopathy+Debrecen+Data+Set#)

```python
import pandas as pd
import numpy as np
from pandas import Series
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
import numpy
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report


#Importação dos dados

a = pd.read_excel('/content/drive/MyDrive/data/a.xlsx')
b = pd.read_excel('/content/drive/MyDrive/data/b.xlsx')
classe=pd.read_excel('/content/drive/MyDrive/data/c.xlsx')
	
#normalização dos dados
def minmax_norm(df_input):
    return (a - a.min()) / ( a.max() - a.min())

df_minmax_norm = minmax_norm(a)

print(df_minmax_norm)

#Juntando os dados no dataframe b

b['a']=df_minmax_norm[['a']]
b['b']=(df_minmax_norm[['b']])
b['c']=(df_minmax_norm[['c']])
b['d']=(df_minmax_norm[['d']])
b['e']=(df_minmax_norm[['e']])
b['f']=(df_minmax_norm[['f']])
b['g']=(df_minmax_norm[['g']])
b['h']=(df_minmax_norm[['h']])
b['i']=(df_minmax_norm[['i']])
b['j']=(df_minmax_norm[['j']])
b['k']=(df_minmax_norm[['k']])
b['l']=(df_minmax_norm[['l']])
b['m']=(df_minmax_norm[['m']])
b['n']=(df_minmax_norm[['n']])
b['o']=(df_minmax_norm[['o']])
b['p']=(df_minmax_norm[['p']])
b['classe']=(classe[['classe']])

# Definindo as entradas (19) e as saidas

entradas= b[b.columns[1:]].to_numpy()
saidas= b[b.columns[0]]

Y=saidas
X=entradas

# Função para criar o modelo
def create_model():
  # cria o modelo
  model = Sequential()
  model.add(Dense(12, input_dim=19, activation='relu'))
  model.add(Dense(8, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  # Compila o modelo
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model


seed = 7
numpy.random.seed(seed)

# criando o modelo com 250 épocas 
model = KerasClassifier(build_fn=create_model, epochs=250, batch_size=10, verbose=1)
# Usando 10-fold para a validação cruzada
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
# Visualização das acurácias dos 10 grupos e da acurácia média
results = cross_val_score(model, X, Y, cv=kfold)
print(results)
print(results.mean())

y_pred = cross_val_predict(model, X, Y, cv=kfold)
conf_mat = confusion_matrix(Y, y_pred)

# Visualização dos valores preditos, da matriz de confusão e do relatório de classificação
print(y_pred)
print(conf_mat)

print(classification_report(Y, y_pred))
```
